```python
import pandas as pd
import numpy as np
import string
import nltk
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import regex as re
from tqdm import tqdm##shows a progession bar when appling function
tqdm.pandas()
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
dataf=pd.read_csv("fake.csv")
datat=pd.read_csv("true.csv")
dataf["label"]=0
datat["label"]=1
```

```python
data=pd.concat([dataf,datat])
```

```python
data=data.drop(["title","subject","date"],axis=1)
##dont run this cell again
```

```python
data= data.sample(frac = 1)
data.reset_index(inplace = True)
data.drop(["index"], axis = 1, inplace = True)
```

```python
manual_data = data.iloc[:50]     
data= data.iloc[50:]


data.head(10)
```

```python
def clean_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return " ".join(tokens)
###need to study about this section, right now is done by chatGPT##
```

```python
data['text'] = data['text'].progress_apply(clean_text)
```

```python
data.head(10)
```

```python
x=data["text"]
y=data["label"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)
```

```python
model=LogisticRegression()
model.fit(xv_train,y_train)
```

```python
y_pred=model.predict(xv_test)
```

```python
model.score(xv_test,y_test)
```

```python
classification_report(y_test,y_pred)
```

```python
import pickle

# Save the trained model
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save the fitted vectorizer
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

with open("manual_news.pkl", "wb")as f:
    pickle.dump(manual_data,f)

```
